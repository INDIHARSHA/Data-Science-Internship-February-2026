### 1. User Login Check
* Problem Statement:
* Given a username and password, check whether login is successful.
* username = "admin"
* password = "1234"


```python
username=input("Please enter the username ! :")
password=input("Please enter the password : ")
if (username =="admin" and password=="1234") :
    print("Login Successful")
# checks if both match the values.
else:
    print("Invalid Credentials")

```

### 2. Pass / Fail Analyzer

### Problem Statement:
Analyze student results.
* marks = [45, 78, 90, 33, 60]


```python
marks = [45, 78, 90, 33, 60]
pass_count=0
for mark in marks:
    if mark>=50: 
        pass_count+=1
print(f"Number of passed students :{pass_count}")
print(f"Number of failed students : {len(marks)-pass_count}")
```

### 3. Simple Data Cleaner

### Problem Statement:
Clean and standardize user names.
* names = [" Alice ", "bob", " CHARLIE "]


```python
names = [" Alice ", "bob", " CHARLIE "]
clean_data=[]
for name in names:
    clean_data.append(name.strip().lower())
clean_data
```

### 4. Message Length Analyzer

### Problem Statement:
Analyze message sizes.
* messages = ["Hi", "Welcome to the platform", "OK"]


```python
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages :
    if len(message)>10:
        print(f"{len(message)} :Flag")
    else:
        print(f"{len(message)}")
```
